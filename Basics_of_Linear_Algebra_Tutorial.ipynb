Matrix Power

```python
a = np.random.random([4,4])

a_2 = np.linalg.matrix_power(a, 2)
a_5 = np.linalg.matrix_power(a, 5)

a_inv_1 = np.linalg.matrix_power(a, -1)
a_inv_5 = np.linalg.matrix_power(a, -5)

print('Matrix \n', a)
print('\n Matrix to the power 2\n', a_2)
print('\n Matrix to the power 7\n', a_5)
print('\n Matrix inverse \n', a_inv_1)
print('\n Matrix cubic inverse \n', a_inv_5)
```

```python
a = np.random.randint(1,10,[4,4])

b = np.array([1,2,3,4])

x = np.linalg.solve(a,b)
print('Coefficient Matrix')
print(a)
print('\nDependent Variable vector')
print(b)
print('\nSolution')
print(x)
```

Singular Value Decomposition

```python
from PIL import Image
import matplotlib.pyplot as plt

image_path = "/content/drive/MyDrive/Machine Learning Minors/f8.jpg"
image = Image.open(image_path)

image_array = np.array(image)

print(image_array.dtype, image_array.min(), image_array.max(), image_array.shape)

plt.imshow(image_array)
plt.axis('off')
plt.show()

img = image_array / 255.0
img = np.transpose(img, axes=(2, 0, 1))  # Shape: (channels, height, width)
```

```python
channels, height, width = img.shape

U_list = []
S_list = []
V_list = []

for i in range(channels):
    U, S, V = np.linalg.svd(img[i], full_matrices=False)
    U_list.append(U)
    S_list.append(S)
    V_list.append(V)

U_array = np.array(U_list)
S_array = np.array(S_list)
V_array = np.array(V_list)

print("U shape:", U_array.shape)
print("S shape:", S_array.shape)
print("V shape:", V_array.shape)
```

```python
Sigma = np.zeros((channels, height, width))

for i in range(channels):
    singular_values = S_array[i, :]
    diag_matrix = np.diag(singular_values)
    Sigma[i, :diag_matrix.shape[0], :diag_matrix.shape[1]] = diag_matrix

print("Sigma shape:", Sigma.shape)

reconst = np.zeros((channels, height, width))

for i in range(channels):
    U_i = U_array[i]
    Sigma_i = np.diag(S_array[i])
    V_i = V_array[i]

    if Sigma_i.shape[1] == V_i.shape[0]:
        reconst[i] = U_i @ Sigma_i @ V_i
    else:
        raise ValueError(f"Dimension mismatch: U_i shape {U_i.shape}, Sigma_i shape {Sigma_i.shape}, V_i shape {V_i.shape}")

reconst = np.transpose(reconst, axes=(1, 2, 0))

plt.imshow(reconst)
plt.axis('off')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

k = 50

reconst = np.zeros((channels, height, width))

for i in range(channels):
    U_i = U_array[i]
    S_i = S_array[i]
    V_i = V_array[i]
    k = min(k, len(S_i))


    Sigma_truncated = np.zeros((U_i.shape[1], k))
    Sigma_truncated[:k, :k] = np.diag(S_i[:k])

    V_truncated = V_i[:k, :]  # Shape: (k, width)

    print(f"Channel {i}:")
    print("U_i shape:", U_i.shape)
    print("Sigma_truncated shape:", Sigma_truncated.shape)
    print("V_truncated shape:", V_truncated.shape)

    reconst[i] = U_i @ Sigma_truncated @ V_truncated

reconst = np.transpose(reconst, axes=(1, 2, 0))

plt.imshow(reconst)
plt.axis('off')  # Turn off axis labels
plt.show()
```
